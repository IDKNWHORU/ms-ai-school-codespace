```python
%pip install openpyxl
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('../data/fine_dust.xlsx', index_col='area')
```

```python
data.head()
```

```python
data2018 = data[2018]
```

```python
data2018
plt.figure(figsize=(15,4))
plt.plot(data2018,color='b',marker='o')
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2018 fine dust line graph')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,4))

for year in range(2016,2020):
    chartdata=data[year]
    plt.plot(chartdata, marker='o', label=year)
    
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2016~2019 fine dust line graph')
plt.legend()
plt.grid()
plt.show()
```

```python
# 막대그래프

data2017 = data[2017]
plt.figure(figsize=(15,4))
plt.bar(data2017.index, data2017,color='g')
plt.ylim(33,55)
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2017 fine dust line graph')
plt.grid()
plt.show()
```

```python
# 산점도 그래프
data2 = pd.read_excel('../data/health_screenings_2020_1000ea.xlsx')
```

```python
heightdata = data2['height']
weightdata = data2['weight']

plt.figure(figsize=(15,4))

plt.scatter(heightdata, weightdata)
```

```python
data2.head()
data2.columns
```

```python
# 수축기 혈압, 이완기 혈압, 공복 혈당 그룹 산점도 그래프
systolic = data2['systolic']
diastolic = data2['diastolic']
blood_sugar = data2['blood_sugar']
plt.figure(figsize=(15,4))

plt.scatter(systolic ,diastolic ,color='r', label='systolic*diastolic')
plt.scatter(systolic ,blood_sugar ,color='g', label='systolic*blood_sugar')
plt.scatter(diastolic ,blood_sugar ,color='b', label='diastolic*blood_sugar')

plt.xlim(40, 100)
plt.ylim(30, 300)
plt.title('2020 Health Screening Group Scatter Graph')
plt.legend()
plt.show()
```

```python
# 여성의 성별, 키 데이터로 여성 키 히스토그램

# 데이터셋 생성
female_data = data2.loc[data2['gender']==2][['gender','height']]
plt.figure(figsize=(10, 6))
plt.hist(female_data['height'], bins=5, label='Female')
plt.xlim(130, 180)
plt.title('2020 Female Health Screening Group Histogram Graph')
plt.legend()
plt.grid()
plt.show()
```

```python
# 그룹 히스토그램 작성
male_data = data2.loc[data2['gender']==1][['gender','height']]
plt.figure(figsize=(10, 6))
plt.hist(female_data['height'], alpha=0.5, bins=5, label='Female')
plt.hist(male_data['height'], alpha=0.5, bins=5, label='male')
plt.xlim(130, 190)
plt.title('2020 Health Screening Group Histogram Graph')
plt.legend()
plt.grid()
plt.show()
```

```python
# 상자수염 그래프
import numpy as np

height = np.array([male_data['height'], female_data['height']],dtype=object)
```

```python
plt.figure(figsize=(10,6))
plt.boxplot(height, labels=['Male', 'Female'])

plt.ylim(130, 200)
plt.title('2020  Health Screening Group box plot Graph')
plt.legend()
plt.show()
```
