## 도시의 위도 검색

```python
cities[['city','latitude']]
```

```python
# 검색할 열만 가져오기
# dframe.loc(조건, 가져올 열이름)
latitude = cities.loc[cities.city=='Abuja', 'latitude'].iloc[0]
```

```python
constellations.loc[(constellations['latitudestart'] >= latitude) & (latitude >= constellations['latitudeend']), 'constellation'].tolist()
```

```python
# 위도 가져오는 사용자 정의 함수
def predict_best_shower_viewing(city) :
    
    if city not in cities.values:
        return city + '는 현재 예측할 수 없습니다'
    
    latitude = cities.loc[cities.city==city, 'latitude'].iloc[0]
    
    constellations_list = constellations.loc[(constellations['latitudestart'] >= latitude) & (latitude >= constellations['latitudeend']), 'constellation'].tolist()
    
    if not constellations_list:
        shower_string = city + "에서는 볼 수 있는 유성우는 없습니다."
        return shower_string
        
    shower_string = city + '에서 유성우를 볼 수 있습니다.'
    return shower_string
    
    return latitude
```

```python
incity = input('city>> ')

print(predict_best_shower_viewing(incity))
```
