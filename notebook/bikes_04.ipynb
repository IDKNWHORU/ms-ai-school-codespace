```python
import bikes_data

bikes = bikes_data.data_prerocessing()
bikes.head(3)
```

### 인기있는 대여소 분석
* 주말에 인기있는 대여소 Top50
* 평일에 인기있는 대여소 Top50

```python
rent_bike = bikes.pivot_table(index=['대여 대여소명', '대여점 위도', '대여점 경도'],
                              columns=['주말구분'],
                              values='자전거번호',
                              aggfunc='count')

rent_bike
```

```python
rent_bike.sort_values('주말', ascending=False).head(50)
```

```python
weekend_house50 = rent_bike.nlargest(50, '주말')[['주말']].reset_index()
weekday_house50 = rent_bike.nlargest(50, '평일')[['평일']].reset_index()
```

```python
weekend_house50
weekday_house50
```

```python
import folium

lat = bikes['대여점 위도'].mean()
lon = bikes['대여점 경도'].mean()
center = [lat, lon]
map1 = folium.Map(location=center, zoom_start = 11)

for i in weekend_house50.index:
    sub_lat = weekend_house50.loc[i, '대여점 위도']
    sub_lon = weekend_house50.loc[i, '대여점 경도']
    name = weekend_house50.loc[i, '대여 대여소명']
    
    folium.Marker(location=[sub_lat, sub_lon],
                  popup=name).add_to(map1)
    
map1
```

```python
for i, row in weekday_house50.iterrows():
    sub_lat = row['대여점 위도']
    sub_lon = row['대여점 경도']
    name = row['대여 대여소명']
    
    folium.Marker(location=[sub_lat, sub_lon],
                  popup=name, icon=folium.Icon(color='red', icon='bicycle', prefix='fa')).add_to(map1)
    
map1
```
