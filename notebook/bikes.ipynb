```python
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
for i in range(1,4):
    temp = pd.read_csv(f"../data/서울특별시 공공자전거 대여정보_201906_{i}.csv", encoding="cp949")
    df = pd.concat([df, temp])
```

```python
df['대여일시']=df['대여일시'].astype('datetime64[ms]')
```

```python
df2 = pd.DataFrame([[4,9],[1,4],[5,6]], columns=['A','B'])

# x 값을 입력받아서 1증가한 후 결과값을 반환
def plus_one(x):
    x+=1
    return x

df2['B'] = df2['B'].apply(plus_one)
```

```python
요일 = ['월', '화', '수', '목', '금', '토', '일']

df['요일'] = df['대여일시'].dt.dayofweek.apply(lambda x : 요일[x])
```

```python
bike_shop = pd.read_csv('../data/공공자전거 대여소 정보_23._06.csv', encoding='cp949')
```

```python
bike_shop.head()
```

```python
bike_shop.columns
bike_gu = bike_shop[['자치구', '대여소\n번호', '보관소(대여소)명', '위도', '경도']]
```

```python
bike_gu = bike_gu.rename(columns={'대여소\n번호': '대여소번호', '보관소(대여소)명': '대여소명'})
```

```python
bikes = pd.merge(df, bike_gu, left_on='대여 대여소번호', right_on='대여소번호')
```

```python
bikes
```

```python
bikes = bikes.drop(['대여소번호', '대여소명'], axis=1)
```
